```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'value1': ['A', 'B', 'C']
})

df2 = pd.DataFrame({
    'id': [2, 3, 4, 5],
    'value2': ['D', 'E', 'F', 'G']
})

mergedFrame2 = pd.concat([df1, df2], axis=1,keys=['DF1','DF2'])
mergedFrame2
```

```python
# Fill NaN values with a specific value
# For example, filling NaN values with 20
# inplace=True modifies the DataFrame directly
mergedFrame2.fillna(20, inplace=True)
mergedFrame2
```

```python
# Fill NaN values with the mean of the column
mergedFrame3 = pd.concat([df1, df2], axis=1, keys=['DF1', 'DF2'])
mergedFrame3.fillna(mergedFrame3.mean(numeric_only=True), inplace=True)
mergedFrame3
```

```python
dataFrame = pd.read_csv('../obesity_prediction.csv')
dataFrame.head()
```

```python
# Example of setting a specific cell to NaN
dataFrame.loc[1, 'Age'] = np.nan
dataFrame.loc[1, 'Height'] = np.nan
dataFrame.loc[1, 'SMOKE'] = np.nan
dataFrame.head()
```

```python
# Fill NaN values in the 'Age' column with 30
# This does not modify the original DataFrame unless we specify inplace=True
dataFrame['Age'] = dataFrame['Age'].fillna(30)
dataFrame.head()
```

```python
# Fill NaN values in multiple columns with different strategies
# Here, we fill 'SMOKE' NaNs with 'no' and 'Height' NaNs with the mean of the 'Height' column
dataFrame.fillna({'SMOKE': 'no', 'Height': dataFrame['Height'].mean()}, inplace=True)
dataFrame.head()
```

```python
# Drop rows with any NaN values
# This will remove any rows that contain at least one NaN value
dataFrame.loc[1, 'Age'] = np.nan
dataFrame.dropna(inplace=True)
dataFrame.head()
```

```python
# Group by
# Grouping by Gender and counting the number of occurrences in each group
dataFrame.groupby(['Gender']).count()
```
