```python
import pandas as pd

# Read data from csv file
newData = pd.read_csv('../obesity_prediction.csv')
newData
```

```python
# Update data
# Use .loc[row_index, col_index] = new_value
# Its update the dataFrame inplace not csv file
newData.loc[0, 'Age'] = 25
newData.loc[1, 'Height'] = 1.75
newData.head()
```

```python
# Update data using .at[row_index, col_index] = new_value
newData.at[2, 'Weight'] = 70
newData.head()
```

```python
# Update full column
newData.loc[:, 'Age'] = 30              # using .loc or shorthand syntax use newData['Age'] = 30
newData.head()
```

```python
# Resetting the DataFrame to original state
newData = pd.read_csv('../obesity_prediction.csv')
newData
```
