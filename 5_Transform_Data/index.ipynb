```python
import pandas as pd

# Read data from csv file
newData = pd.read_csv('../obesity_prediction.csv')
newData.head()
```

```python
# Transforming Data using .apply() method
# It applies a function along an axis of the DataFrame (rows or columns)
# Here we will increase the age of each person by 5 years
def increase_age(age):
    return age + 5

# .apply returns a new Series or DataFrame with the transformed data, Need to add it back to the DataFrame
newData['Age'] = newData['Age'].apply(increase_age)
newData.head()
```

```python
# Alternatively, using a lambda function
# lambda function is an anonymous function defined using the lambda keyword
newData['Age'] = newData['Age'].apply(lambda x: x + 5)  # Using lambda function to increase age by 5
newData.head()
```

```python
newData['Height'] = newData['Height'].apply(lambda x: x + 1)  # Increase height by 1 inch
newData.head()
```
