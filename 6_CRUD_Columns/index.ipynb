```python
import pandas as pd

# Read data from csv file
newData = pd.read_csv('../obesity_prediction.csv')
newData.head()
```

```python
# Inserted new column
newData['New_Column'] = 'New Value'
newData.head()
```

```python
# Inserted new column based on condition
# If age < 25 then 'Young' else 'Adult'
newData['Age_Category'] = ['Young' if age < 25 else 'Adult' for age in newData['Age']]
newData.head()
```

```python
# Insert new column on specific location
# .insert(index, column_name, value)
newData.insert(2, 'BMI', newData['Weight']/newData['Height']**0.02)  # 2 is the index position where we want to insert the new column
newData.head()
```

```python
# Delete column
# newData.drop(columns=['New_Column', 'BMI'], inplace=True)
# inplace=True to make the changes in the original DataFrame, default is False
newData.drop(columns=['New_Column', 'BMI'], inplace=True)
newData.head()
```

```python
# Delete column
# Use del keyword
del newData['Age_Category']
newData.head()
```

```python
# Rename column
# .rename(columns={'old_name': 'new_name'}, inplace=True)
# inplace=True to make the changes in the original DataFrame, default is False
newData.rename(columns={'Weight': 'Weight_kg', 'Height': 'Height_m'}, inplace=True)
newData.head()
```
