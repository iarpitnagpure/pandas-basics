```python
# Display summary statistics of the DataFrame including count, mean, std, min, max, and quartiles
newData.describe()
```

```python
# Location based indexing with .loc
# Display the values of the 'Age', 'Height', and 'Weight' columns at first row (index 0)
# Use .loc to access the row and column .loc[row_index, column_label]
newData.loc[0, ['Age', 'Height', 'Weight']]
```

```python
# Display the values of the 'Age', 'Height', and 'Weight' columns for the first 10 rows
# In slicing last row & column is also inculded
newData.loc[0:10, 'Age':'FAF']
```

```python
# Display all rows
newData.loc[:, 'Age':'Weight']
```

```python
# Integer based indexing with .iloc
# Display the values of the first three columns at the first row (index 0), Column must be integer
newData.iloc[0, 0]
```

```python
# Slicing with .iloc to get first 3 rows and first 3 columns
# In slicing last row & column is excluded (Only with .iloc)
newData.iloc[0:3, 0:3]
```

```python
# Accessing single value using .at (Slicing is not allowed with .at)
# Display the value of the 'Age' column at the first row (index 0)
newData.at[0, 'Age']
```

```python
# Accessing single value using .iat (Slicing is not allowed with .iat)
# Display the value at the first row (index 0) and first column (index 0), Column must be integer
newData.iat[0, 0]
```
